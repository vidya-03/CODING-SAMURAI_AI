```python
import re
```

```python
def cleanTxt(text):
    text = str(text)
    text = text.lower()
    text = re.sub(r'[;:!\\"\'()\[\]]',"", text)
    text = re.sub(r'(<br\s*/><br\s*/>)|(\-)|(\/)'," ", text)
    return text
```

```python
cleanTxt("Basically there's a family where a little boy (Jake) thinks there's a zombie in his closet & his parents are fighting all the time.<br /><br />This movie is slower than a soap opera... and suddenly, Jake decides to become Rambo and kill the zombie.<br /><br />OK, first of all when you're going to make a film you must Decide if its a thriller or a drama! As a drama the movie is watchable. Parents are divorcing & arguing like in real life. And then we have Jake with his closet which totally ruins all the film! I expected to see a BOOGEYMAN similar movie, and instead i watched a drama with some meaningless thriller spots.<br /><br />3 out of 10 just for the well playing parents & descent dialogs. As for the shots with Jake: just ignore them.")
```

```python
##Using CountVectorizer
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vectorizer_pos = CountVectorizer(tokenizer=nltk.word_tokenize,min_df=1000)
```

```python
count_vectorizer_neg = CountVectorizer(tokenizer=nltk.word_tokenize,min_df=1000)
```

```python
count_vectorizer_pos.fit(pos['review'])
```

```python
count_vectorizer_neg.fit(neg['review'])
```

```python
dict_most_common_pos=count_vectorizer_pos.vocabulary_
```

```python
dict_most_common_neg=count_vectorizer_neg.vocabulary_
```

```python
most_common_pos=list(dict_most_common)
```

```python
most_common_pos
```

```python
most_common_neg=list(dict_most_common_neg)
```

```python
most_common_neg
```

```python
def feature(sent):
    count_neg = 0
    count_pos = 0
    word = nltk.word_tokenize(sent)
    for i in word:
        if i in most_common_pos[:500]:
            count_pos+=1
        if i in most_common_neg[:500]:
            count_neg+=1
    feat={'positive_words':count_pos, 'negative_words':count_neg}
    return feat
```

```python
feature(pos.iloc[1,0])
```

```python
featureset=df['review'].apply(feature)
```

```python
featureset=featureset.to_list() #convert from object type to list
```

```python
featureset
```

```python
featuredata=pd.DataFrame(featureset)
```
